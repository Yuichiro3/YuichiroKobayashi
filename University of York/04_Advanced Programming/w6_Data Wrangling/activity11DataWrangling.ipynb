```python
import pandas as pd
```

```python
# Load CSV file
filename = "CollegeGrades1.csv"
college_grades_df = pd.read_csv(filename)
```

```python
# Rename the columns
college_grades_df.columns = [
    'Label', 'Type', 'Art_F', 'Art_P', 'Art_M', 'Art_D',
    'Eng_F', 'Eng_P', 'Eng_M', 'Eng_D',
    'Psy_F', 'Psy_P', 'Psy_M', 'Psy_D',
    'Math_F', 'Math_P', 'Math_M', 'Math_D',
    'Chem_F', 'Chem_P', 'Chem_M', 'Chem_D',
    'Tech_F', 'Tech_P', 'Tech_M', 'Tech_D',
    'Extra_1', 'Extra_2', 'Extra_3', 'Extra_4', 'Extra_5', 
    'Extra_6', 'Extra_7', 'Extra_8', 'Extra_9', 'Extra_10',
    'Extra_11', 'Extra_12'
]
college_grades_df
```

```python
# Filter only the rows relevant to "Year 1", "Year 2", "Evening"
college_grades_cleaned_df = college_grades_df[
    college_grades_df['Label'].isin(['Year 1', 'Year 2', 'Evening']) # Filterling 
].copy()

college_grades_cleaned_df
```

```python
# Drop unnecessary columns
college_grades_cleaned_df = college_grades_cleaned_df[
    [
        'Label', 'Art_F', 'Art_P', 'Art_M', 'Art_D',
        'Eng_F', 'Eng_P', 'Eng_M', 'Eng_D',
        'Psy_F', 'Psy_P', 'Psy_M', 'Psy_D',
        'Math_F', 'Math_P', 'Math_M', 'Math_D',
        'Chem_F', 'Chem_P', 'Chem_M', 'Chem_D',
        'Tech_F', 'Tech_P', 'Tech_M', 'Tech_D'
    ]
]

college_grades_cleaned_df
```

```python
# Convert columns to a multi-level structure with subjects as parents and grades as sub-columns
college_grades_cleaned_df.columns = pd.MultiIndex.from_tuples([
    ('Year', ''),
    ('Art', 'F'), ('Art', 'P'), ('Art', 'M'), ('Art', 'D'),
    ('Engineering', 'F'), ('Engineering', 'P'), ('Engineering', 'M'), ('Engineering', 'D'),
    ('Psychology', 'F'), ('Psychology', 'P'), ('Psychology', 'M'), ('Psychology', 'D'),
    ('Mathematics', 'F'), ('Mathematics', 'P'), ('Mathematics', 'M'), ('Mathematics', 'D'),
    ('Chemistry', 'F'), ('Chemistry', 'P'), ('Chemistry', 'M'), ('Chemistry', 'D'),
    ('Technology', 'F'), ('Technology', 'P'), ('Technology', 'M'), ('Technology', 'D')
])

college_grades_cleaned_df
```
