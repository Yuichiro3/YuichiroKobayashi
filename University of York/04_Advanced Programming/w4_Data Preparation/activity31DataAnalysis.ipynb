```python
import numpy as np
import pandas as pd
```

```python
# Load the csv
fileName = "PeoplesFavourites-1.csv"
df = pd.read_csv(fileName, header=0)
df.head()
```

```python
# Set Unique ID
df["Unique ID"] = ["People" + str(i) for i in range(1, len(df)+1)]
df
```

```python
# Remove the first and last name
df.drop(columns=["First Name", "Last Name"], inplace=True)
df
```

```python
# Set null for missing values
null_symbols = ["", " ", "NA", "N/A", "-", "null", "none"]
df.replace(null_symbols, np.nan, inplace=True)
df
```

```python
# Delete the rows that have NaN in "Email"
df.dropna(subset=["Email"], inplace=True)
df
```

```python
# Export the new file
df.to_csv('Cleaned_PeoplesFavourites-1.csv', index=False)
```
