```python
import numpy as np
import pandas as pd
```

```python
# Load the csv
fileName = "SalesData.csv"
df = pd.read_csv(fileName)
```

```python
# See the first five data
df.head()
```

```python
# Rename unnamed column
df.rename(columns={'Unnamed: 0': 'Store'}, inplace=True)
df.set_index("Store", inplace=True)
df
```

```python
# Calculate the average sales for each store across the entire year
store_averages = df.mean(axis=1)
store_averages
```

```python
# Create a new DataFrame to store the percentage increase for each month
percentage_increase_df = df.apply(lambda x: (x-store_averages) / store_averages * 100, axis=0)
```

```python
# Find the store with the largest and smallest increase and the corresponding months
largest_increase = percentage_increase_df.stack().idxmax()
smallest_increase = percentage_increase_df.stack().idxmin()
largest_increase_value = percentage_increase_df.stack().max()
smallest_increase_value = percentage_increase_df.stack().min()
```

```python
# Calculate the average percentage increase across all stores for each month
monthly_average_increase = percentage_increase_df.mean(axis=0)

# Find the months with the smallest and largest average increase across all stores
smallest_avg_increase_month = monthly_average_increase.idxmin()
largest_avg_increase_month = monthly_average_increase.idxmax()
smallest_avg_increase_value = monthly_average_increase.min()
largest_avg_increase_value = monthly_average_increase.max()
```

```python
# Display the results
print("Percentage Increase Data Table (based on store average for the year):\n", percentage_increase_df)
print("\nStore with the largest increase:")
print(f"Store: {largest_increase[0]}, Month: {largest_increase[1]}, Increase: {largest_increase_value}%")
print("\nStore with the smallest increase:")
print(f"Store: {smallest_increase[0]}, Month: {smallest_increase[1]}, Increase: {smallest_increase_value}%")
print("\nAcross all stores:")
print(f"Largest average monthly increase: {largest_avg_increase_value}% in {largest_avg_increase_month}")
print(f"Smallest average monthly increase: {smallest_avg_increase_value}% in {smallest_avg_increase_month}")
```
