```python
import numpy as np
```

```python
# Define the array
arr = np.array([3, 9, 7, 4, 3, 7, 8, 1, 2, 4, 5, 6, 3, 2, 1, 4, 5, 7, 7, 4, 2, 1, 6, 7])
```

```python
# Use the sort algorithm (Mergesort)
sorted_arr_merge = np.sort(arr, kind="mergesort")
print(sorted_arr_merge)
```

```python
# Use the sort algorithm (Heapsort)
sorted_arr_heap = np.sort(arr, kind="heapsort")
print(sorted_arr_heap)
```
