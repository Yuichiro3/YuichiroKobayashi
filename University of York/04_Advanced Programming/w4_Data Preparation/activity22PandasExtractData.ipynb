```python
import numpy as np
import pandas as pd
```

```python
# Load the csv
fileName = "SalesData.csv"
df = pd.read_csv(fileName)
```

```python
# See the first five data
df.head()
```

```python
# Rename unnamed column
df.rename(columns={'Unnamed: 0': 'Store'}, inplace=True)
df.set_index("Store", inplace=True)
df
```

```python
# Sales for P2 and B8 in Nov-18, Feb-18, and Mar-19
sales_p2_b8_df = df.loc[["P2","B8"], ["Nov-18", "Feb-19", "Mar-19"]]
sales_p2_b8_df
```

```python
# Third quarter sales data for London stores (L3 and L1) from Oct-18 to Dec-18
london_sales_q3 = df.loc[["L3", "L1"], ["Oct-18", "Nov-18", "Dec-18"]]

# Calculate the monthly percentage increase for each store in London (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html)
london_sales_pct_increase = london_sales_q3.pct_change(axis=1) * 100  

# Display the third quarter sales data and the percentage increase
print("Third quarter sales for London (L3 & L1):\n", london_sales_q3)
print("\nMonthly percentage increase for London (L3 & L1):\n", london_sales_pct_increase)
```

```python
# Extract data for New York stores (N4 and N6)
ny_sales_df = df.loc[["N6", "N4"]]
ny_sales_df
```

```python
# Stack the data to transform columns (months) into a single column, making it easier to find top values
top_three_ny_sales = ny_sales_df.stack().nlargest(3) # stack is used to conver the DataFrame to a Series
top_three_ny_sales
```

```python
# Find the overall lowest sales figure along with the store and month it occured
lowest_sales = df.stack().idxmin() # Find the index (store and month) where the minimum value occures
lowest_sales_value = df.stack().min()

print(f"The overall lowest sales figure is {lowest_sales_value}, which occurred in {lowest_sales[0]} in {lowest_sales[1]}")
```
