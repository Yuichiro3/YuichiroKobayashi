```python
import pydp as dp  # PyDPはGoogleのC++ DPライブラリのPythonラッパー
from pydp.algorithms.laplacian import BoundedSum
```

```python
# 数値データ（例：ユーザーの年齢など）
data = [23, 25, 32, 45, 31, 38, 50, 28, 22]

# 差分プライバシー付き平均
mean = dp.algorithms.laplacian.BoundedMean(epsilon=1.0, lower_bound=0, upper_bound=100)
for val in data:
    mean.add_entry(val)

private_mean = mean.result()
print(f"差分プライバシーを適用した平均: {private_mean}")
```

```python
data = [30, 45, 60, 90, 120]  # 1日の使用時間（分）

dp_sum = BoundedSum(epsilon=8.0, lower_bound=0, upper_bound=180)
for val in data:
    dp_sum.add_entry(val)

private_sum = dp_sum.result()
print(f"プライベートな合計: {private_sum}")
```
